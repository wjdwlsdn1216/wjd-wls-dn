 그 후 많은 연구의 시도들이 있었지만 1970년대  컴퓨터 성능의 한계로 성과가 미미해지자 투자가 중단되어 첫번째 암흑기가 도래하고, 수 년이 지난후 전문가 시스템개발과함께 다시 주목되고 인터넷 검색엔진 등으로 활용되었지만 유용성 비판과함게 두번째 암흑기가 시작된다.

- **2-1 음성인식 인공지능**

음성인식 기술은 1952년에 연구가 시작되어 현재까지 5세대에 걸쳐 발
전되어 왔으며 최근 딥러닝 기술과 결합되어 더욱 활발한 연구가 진행중에 있다.

음성인식 기술은 소프트웨어개발이 전부라해도 과언이 아니다. 음성인식 소프트웨어란 외부에서 발생된 음성을 텍스트와 같은 정보로 변환하고 변환된 정보를 스스로 이해하며, 한편으로 정보를 합성음성으로 변환해주는 등 기술을 통해 제품이 사용자의 요구에 따라 그에 맞게 적합한 대답을 제공하는 시스템 소프트웨어이다.

음성인식 소프트웨어는 관점에 따라 다양하게 분류할 수 있다. 발화자에 따라 화자종속방식, 화자독립방식, 화자적응방식으로 구분되고, 발음에 따라 고립단어, 핵심어, 연결단어, 연속음성, 대화음성으로 분류할 수 있고 인식대상어휘 수에 따라 소, 중, 대용량 인식기술로 나누어 지며, 또 단어단위로 인식하는 단어인식기술, 음소단위로 인식하는 가변어인식기술로도 나눌 수 있다.

음성인식 기술의 처리 과정을 보면 먼저 사람이 음성인식을 하고자 하는 문장을 발성하면 해당 문장에 대해 음성구간에 해당하는 시작점과 끝점을 검출하고  음성 구간을 검출하면 특징벡터를 추출하고 추출된 특징벡터를 기반으로 잡음 제거 및 채널보상 과정을 거치고 이러한 과정을 거친 입력 특징 벡터는 음향모델과 언어모델을 사용하여 인식 대상인 연속 어휘 패턴들을 효율적으로 비교하는 탐색과정이다. 이 과정은 단어 레벨과 문장 레벨의 패턴 정보가 서로 결합하여 최종적인 인식 문장을 찾아낼 수 있도록 탐색공간을 구성한 후비터비 탐색 기법을 사용하여 이루어진다. 즉 언어모델을 적용해 시간에 따라 어휘열을 찾는 것이 단
어탐색이고, 이렇게 시간별로 가능성이 있는 어휘들의 다양한 조합을 통해 최종 결과를 생성하는 것이문장탐색 비교이다. 마지막으로 문장레벨 탐색 결과에 대해서 수락 또는 거절을 결정하는 발화검증으로 인식시스템이 이루어져있다.

그렇다면 이러한 음성인식 기술을 가지고있는 제품을 알아보자. 음성인식 기술을 활용한 제품중 당연 으뜸은 음성인식 스피커라 할 수 있다. 많은 음성인식 스피커중 아마존의 에코는 음성인식 기술과 스피커를 결합해 스포티파이 판도라등을 통해 음악을 들려주고, 오디오 북을 들려줄 뿐만 아니라, 뉴스나 교통 상황이나 날씨 스포츠 뉴스 등을 사용자에게 알려주며 위모 필립스 휴 삼성 스마트씽즈 네스트 등과 연계된 서비스를 제공한다. 에코는 아직 완성도 높은 가상비서 수준에 도달하지는 못했지만 최소한 라디오의 대체재로는 충분하다고 생각한다.

- **2-2 언어분야 인공지능**

언어분야에 관련한 인공지능은 여러가지가 있겠지만 그중 자연어 대화 인터페이스 즉, 인간의 언어를 기계에서 처리하는 기술을 일반적으로 자연어 처리라고 한다.우리가 평소에 자주 사용하는 구글 네이버 다음 검색등이 자연어 처리기술을 이용한 검색 서비스이다. 최근에는 빅데이터 분석, 대화 인터페이스, 번역기 등 다양한 서비스와 어플에서 사용되고 있다. 자연어 대화 인터페이스에서 처리하는 대화 유형은 크게 사회적 습관이나 지식, 대화모델을  바탕으로 의사소통을 위한 채팅형 대화와 사용자의 발화 의도를 이해하고 발화의 목적의 따라 적합한 발화 또는 작업을 수행하는 목적 지향형 대화로 나눌 수 있다.

자연어 대화 인터페이스기술 처리 과정을 살펴보면 첫번째로 발화 이해, 대화 관리, 응답 생성으로 나뉘어진다. 발화 이해는 텍스트 형태의 발화를 대상으로 사용자의 의도를 연어 대화 인터페이스기술 처리 과정을 살펴보면 첫번째로 발화 이해, 대화 관리, 응답 생성으로 나뉘어진다. 발화 이해는 텍스트 형태의 발화를 대상으로 사용자의 의도를 분석하고 파악하며 발화 의미구조를 생성하는 단계이고, 대화 관리는 채팅 대화 모델을 이용하여 응답 의미구조를 생성하는 단계이며, 응답생성은  그동안 학습해온 지식과 서비스 수행엔진을 이용하여 학습된 내용을 바탕으로 적절한 응답을 생성하는 단계이다.

2018년 카카오는 24시간 상담로봇 챗봇을 출시했다. 기본적인 상담 업무를 할 뿐만 아니라 금융 용어를 설명해주고 예금,적금, 대출 상환 금액, 해외 송금 환율 등의 계산기까지 탑재해 많은 이들에게 호평받았다. 그리고 현재는 약 1만7천개의 카카오톡 챗봇이 운용되고 있다. 24시간 고객 상담이 가능한 상담 챗봇, 원격으로 음료수를 주문 결제하는 챗봇 주문, 쇼핑몰 구매 배송 내역을 조회하고 상품 검색과 구매까지 할 수 있는 챗봇 쇼핑 서비스가 대표적으로 있다.
